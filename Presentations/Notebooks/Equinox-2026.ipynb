## Wolfram Language interaction

Wolfram Language -- formerly Mathematica -- can be used in Raku sessions via ["Proc::ZMQed"](https://raku.land/zef:antononcube/Proc::ZMQed).  
(See the video ["Using Wolfram Engine in Raku sessions"](https://www.youtube.com/watch?v=nWeGkJU3wdM) for more examples and details.)

```raku
use Proc::ZMQed;

# Make object
my Proc::ZMQed::Mathematica $wlProc .= new(url => 'tcp://127.0.0.1', port => '5550');

# Start the process (i.e. Wolfram Engine)
$wlProc.start-proc;
```

```raku
my $cmd = 'Expand[(x+y)^4]';
my $wlRes = $wlProc.evaluate($cmd);
say "Sent : $cmd";
say "Got  :\n $wlRes";
```

```raku
$wlProc.evaluate('FindAstroEvent["SpringEquinox", Yesterday]');
```

```raku
my $cmd=q:to/END/;
p= GeoGraphics[{GeoStyling["SatelliteNight"], NightHemisphere[]}, GeoProjection -> "Orthographic", GeoCenter -> 45]
ps = ExportString[p,{"Base64","PNG"}]; Export["/tmp/wlimg.txt",ps]
END
```

```raku
$wlProc.evaluate($cmd);
my $timg = slurp('/tmp/wlimg.txt');
from-base64($timg)
```
